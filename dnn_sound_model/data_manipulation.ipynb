```python
import pandas as pd
import numpy as np
import ast

df = pd.read_csv("final_dataset.csv")
freqs = ast.literal_eval(df["freqs[Hz]"][0])
df_train = pd.read_csv("training_data.csv")
df_train
```

```python
# Create a copy of df_train
df_train_mirror = df_train.copy()
# Mirror the zeta values
df_train_mirror['zeta'] = 2 * np.pi - df_train_mirror['zeta']
# Concatenate the original and mirrored DataFrames
df_train_extended = pd.concat([df_train, df_train_mirror], ignore_index=True)
# Calculate the average RPM and replace the columns RPM1, RPM2, RPM3, and RPM4 with RPM
df_train_extended['RPM'] = df_train_extended[['RPM1', 'RPM2', 'RPM3', 'RPM4']].mean(axis=1)
df_train_extended.drop(columns=['RPM1', 'RPM2', 'RPM3', 'RPM4'], inplace=True)
df_train_extended.drop(columns={
    "C_proc",
    "pitch",
    "roll",
    "yaw",
}, inplace=True)
df_train_extended.to_csv("training_data_extended.csv", index=False)
```
